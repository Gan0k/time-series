```r
library(forecast)
library(gdata)
```

```r
data <- read.csv("Data/2006.csv",header = TRUE, sep = ";")
#transform data in time series
n<- dim(data)[1]
n<-24*60
myTs <- ts(data[,4])
#test <- ts(data[301:345,2], start=c(301), end=c(345))
```

```r
mod <- auto.arima(myTs)
summary(mod)
```

```r
fit <- fitted(mod)
```

```r
a <-1000
b <-1110
plotTs4 <- ts(data[a:b,4])
plotFit4 <- ts(fit[a:b])
plot(plotTs4,col="blue")
lines(plotFit4,col="red")
```

```r
kTs <- ts(data[,8])
```

```r
modK <- auto.arima(kTs)
summary(modK)
```

```r
fitK <- fitted(modK)
```

```r
a <-1000
b <-1110
plotTs <- ts(data[a:b,8])
plotFit <- ts(fitK[a:b])
plot(plotTs,col="blue")
lines(plotFit,col="red")
```
